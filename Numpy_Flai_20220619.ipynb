```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
n = 100
x = np.random.uniform(1,10,n)
y = x + np.random.normal(0,1,n)
y2 = -x + np.random.normal(0,1,n)
```

```python
plt.plot (x,y, ".", color  = "green", marker= "*")
plt.show()
```

```python
plt.plot (x,y, ".", color  = "green", marker= "s", alpha = 0.25)
plt.plot(x, y2,".", color = "blue", marker= "^", markersize = 12)
plt.show()
```

```python
import scipy.stats as ss
```

```python
x = np.arange(-4,4,1)
y= ss.norm.pdf(x)
```

```python
plt.plot(x,y, "-")
plt.savefig("normallpadrao.pdf")
```

```python
x = np.arange(-4,4,0.05)
y1= ss.norm.pdf(x, 0, 1)
y2= ss.norm.pdf(x, 0, 2)
y3= ss.norm.pdf(x, 0, 3)
y4= ss.norm.pdf(x, 0, 4)

plt.plot(x, y1, "-", color = "blue")
plt.plot(x, y2, "-", color = "red")
plt.plot(x, y3, "-", color = "yellow")
plt.plot(x, y4, "-", color = "green")
```

```python
x = np.arange(-4,4,0.05)
y1= ss.norm.pdf(x, 0, 1)
y2= ss.norm.pdf(x, 0, 2)
y3= ss.norm.pdf(x, 0, 3)
y4= ss.norm.pdf(x, 0, 4)

plt.plot(x, y1, "-", color = "blue")
plt.plot(x, y2, "-", color = "red")
plt.plot(x, y3, "-", color = "yellow")
plt.plot(x, y4, "-", color = "green")

plt.fill_between(x, y1, color = "blue", alpha = 0.2)
plt.show()
```

```python
plt.fill_between(x, y1, color = "blue", alpha = 0.2)
plt.fill_between(x, y2, color = "red", alpha = 0.2)
plt.fill_between(x, y3, color = "yellow", alpha = 0.2)
plt.fill_between(x, y4, color = "green", alpha = 0.2)
plt.axis('off')
plt.show()

```

```python
##Projeto várias curvas 

p1 = ss.norm.cdf(1) - ss.norm.cdf(-1)
p2 = ss.norm.cdf(1) - ss.norm.cdf(-2)
p3 = ss.norm.cdf(1) - ss.norm.cdf(-3)

print('P(-1 < Z < 1 = {:2.2f}%'.format(100*p1))
print('P(-1 < Z < 2 = {:2.2f}%'.format(100*p2))
print('P(-1 < Z < 3 = {:2.2f}%'.format(100*p3))
```
